```python
import matplotlib.pyplot as plt
```

## Plot

```python
import numpy as np

x = np.array([1, 4, 7, 10])
y = np.array([1, 10, 4, 10])
x, y
```

```python
a = np.array([2, 3, 6, 9])
b = np.array([1, 11, 5, 2])
a, b
```

```python
plt.plot(x, y, marker='o', color='magenta', ls='-.')
plt.xlabel('X Label')
plt.ylabel('Y Label')
plt.title('My plot')
plt.show()
```

```python
plt.plot(x, y, marker='o', color='magenta', ls='-.')  # First line
plt.plot(a, b, color='green')  # Second line
plt.xlabel('X Label')
plt.ylabel('Y Label')
plt.title('My plot')
plt.axis([0, 13, 0, 15])
plt.savefig('my_plot.png')
plt.show()
```

## Sub-Plot

```python
# fig, (pl1,  pl2) = plt.subplots(1, 2)

plt.figure(figsize=(20, 10))

plt.subplot(1, 2, 1)  # numrows=1, numcolumns=2, subplot=1
plt.xlabel('X label')
plt.ylabel('Y label')
plt.title('First one')
plt.plot(x, y, color="cyan")

plt.subplot(1, 2, 2)  # numrows=1, numcolumns=2, subplot=2
plt.xlabel('a label')
plt.ylabel('b label')
plt.title('second one')
plt.plot(a, b, color="green")

plt.show()
```

## Bar-Chart

```python
import pandas as pd
df = pd.DataFrame(columns=['Model', 'Category', 'Color', 'Price'],
                  data=[['Yaris', 'Tyota', 'white', '120'],
                        ['Camry', 'Tyota', 'red', '110'],
                        ['Prius', 'Tyota', 'green', '90'],
                        ['Sienna', 'Tyota', 'white', '115'],
                        ['Camry', 'Tyota', 'blue', '130'],
                        ['Avalon', 'Tyota', 'black', '65'],
                        
                        ['Series1', 'BMW', 'white', '220'],
                        ['Series3', 'BMW', 'black', '260'],
                        ['Series5', 'BMW', 'blue', '460'],
                        ['Series7', 'BMW', 'black', '550'],
                        
                        ['Edge', 'Ford', 'white', '180'],
                        ['Escape', 'Ford', 'pink', '175'],
                        ['F-150', 'Ford', 'white', '150'],
                        
                        ['Ibiza', 'Seat', 'black', '80'],
                        ['Leon', 'Seat', 'black', '88'],
                        ['Leon', 'Seat', 'white', '88'],
                        ['Cupra', 'Seat', 'black', '98'],
                        ['Cupra', 'Seat', 'black', '98'],
                        ['Mii', 'Seat', 'purple', '98'],
                        ['Mii', 'Seat', 'red', '80'],
                        ['Mii', 'Seat', 'yellow', '82'],
                       ])
df
```

```python
a = df['Category'].value_counts()
a
```

```python
plt.title('Bar Chart')
plt.xlabel('Category')
plt.ylabel('Count')
plt.bar(a.index, a.values)
plt.show()
```

## Pie-chart

```python
total = df.shape[0]

perc = [a.iloc[0]/total,
       a.iloc[1]/total,
       a.iloc[2]/total,
       a.iloc[3]/total]
perc
```

```python

plt.title('PIE-Chart')
plt.axis('equal')
explode = (0, 0, 0, 0.1)
plt.pie(perc, labels=a.index, autopct='%1.2f%%', startangle=90, explode=explode)
plt.show()
```
